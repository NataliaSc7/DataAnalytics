```python
import pandas as pd

correlation_data={
    'arppu':[43, 47, 47, 46, 44, 47, 45, 45],
    'paid_users':[100, 111, 145, 165, 179, 199, 188, 189],
    'revenue':[4339, 5171, 6792, 7617, 7922, 9344, 8425, 8440],
    'new_users':[40, 27, 44, 50, 38, 41, 27, 32], 
    'new_revenue':[1573, 985, 1392, 1991, 1406, 1468, 863, 1268],
    'churned_users':[19, 26, 22, 41, 46, 46, 71, 64],
    'churned_revenue':[630, 980, 880, 1519, 1362, 1734, 2800, 1947],
    'returned_users':[6, 10, 12, 11, 22, 25, 33, 33],  
    'returned_revenue':[244, 605, 498, 389, 615, 1046, 1146, 1361],
    'construction_mrr':[1052, 928, 1197, 1586, 1740, 1801, 2291, 2381],
    'expansion_mrr':[604, 1149, 1808, 1550, 1385, 2443, 2164, 1713]}
                        
correlation_df=pd.DataFrame(correlation_data)
correlation_df
```

```python
import pandas as pd
import numpy as np

correlation_matrix = correlation_df.corr()

def highlight_revenue_column(s):
    return ['background-color: lime' if s.name == 'revenue' else '' for v in s]

styled_matrix = correlation_matrix.style.format("{:.2f}").apply(highlight_revenue_column, axis=0)

styled_matrix
```

**Summaries**

-The **aeppu value** demonstrates _the little correlation_ to **revenue** and paid users, churned revenue and roughly the same relation to revenue-type and users-type predicates  - these predicates can be seen as multicollinear and interchangeable;

-**the revenue** demonstrates _the strong correlation_ to **churned users, returned users, construction mrr and expansion mrr** - paid users, churned revenue and returned revenue are not considered due to direct formulae for above mentione;.

-_the weak correlation_ of  **revenue** to **new users** is unexpected, that's why statistical testing is desirable:

```python
from scipy.stats import spearmanr

revenue = correlation_df['revenue']
new_users = correlation_df['new_users']

corr_coefficient, p_value = spearmanr(revenue, new_users)

print(f"correllation coef : {corr_coefficient}")
print(f"p-value: {p_value}")
```

-the **Spearman test** confirms _statistically insignificant_ impact of new users to revenue - it's possible that we just have a small sample and more data may be needed to find meaningful relationships;
